6. Task

```python
import math, cmath

R = 50
L = 0.1
C = 100e-6
f = 60

w = 2 * math.pi * f

Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))

Z_total = Z_R + Z_L + Z_C

print("Total Impedance:", Z_total)
print("Resistive Part:", Z_total.real)
print("Reactive Part:", Z_total.imag)
print("Magnitude:", abs(Z_total))
print("Phase Angle (radians):", cmath.phase(Z_total))
```

7. Task

```python
import cmath
import math

R = 50
L = 0.1
C = 100e-6
f = 60

w = 2 * math.pi * f

Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))

Z_total = Z_R + Z_L + Z_C
V = complex(120, 0)
I = V / Z_total

print(f"Total Impedance: {Z_total.real:.2f} + j{Z_total.imag:.2f} Ω")
print(f"Impedance Magnitude: {abs(Z_total):.2f} Ω")
print(f"Impedance Phase Angle: {math.degrees(cmath.phase(Z_total)):.2f}°")

print(f"\nCircuit Current: {I.real:.2f} + j{I.imag:.2f} A")
print(f"Current Magnitude: {abs(I):.2f} A")
print(f"Current Phase Angle: {math.degrees(cmath.phase(I)):.2f}°")

```

8. Task: Run and understand  

a. Ohm's law

```python
V = 230
I = 5
R = V / I

print(f"The voltage is {V} V, the current is {I} A, so the resistance is {R:.2f} Ω.")
print("The voltage is {} V, the current is {} A, so the resistance is {:.2f} Ω.".format(V, I, R))
```

b. AC power

```python
P = 2000
V = 220
I = P / V

print(f"The real power is {P} W at {V} V supply, drawing {I:.2f} A current.")
print("The real power is {0} W at {1} V supply, drawing {2:.2f} A current.".format(P, V, I))
```

c. RLC Circuit Impedance 

```python
R = 50
X_L = 30
X_C = 20
Z = complex(R, X_L - X_C)

print(f"Impedance of the circuit is {Z.real} + j{Z.imag} Ω.")
print("Impedance of the circuit is {0} + j{1} Ω.".format(Z.real, Z.imag))
```
