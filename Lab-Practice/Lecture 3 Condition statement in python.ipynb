# <span style="font-family:Times New Roman">Understanding Conditional Statements in Python</span>

<span style="font-family:Times New Roman">
In a Python program, the <b>if statement</b> is how you perform decision-making.  
It allows for conditional execution of a statement or a group of statements based on the value of an expression.
</span>

<span style="font-family:Times New Roman">
The outline of this tutorial is as follows:
</span>

1. <span style="font-family:Times New Roman"><b>Quick Overview</b>  
   First, you’ll get a quick overview of the <code>if</code> statement in its simplest form.</span>  

2. <span style="font-family:Times New Roman"><b>Grouping Statements</b>  
   Next, using the <code>if</code> statement as a model, you’ll see why control structures require some mechanism for grouping statements together into compound statements or blocks. You’ll learn how this is done in Python.</span>  

3. <span style="font-family:Times New Roman"><b>Complex Decisions</b>  
   Lastly, you’ll tie it all together and learn how to write complex decision-making code.</span>  


# <span style="font-family:Times New Roman">Introduction to the if Statement</span>

<span style="font-family:Times New Roman">
We’ll start by looking at the most basic type of <b>if statement</b>.  
In its simplest form, it looks like this:
</span>

```python
if <expr>:
    <statement>


````python
x = 0
y = 5

if x < y:   # Truthy
    print("yes")

if y < x:   # Falsy
    print("yes")

if x:   # Falsy
    print("yes")

if y:   # Truthy
    print("yes")

if x or y:   # Truthy
    print("yes")

if x and y:   # Falsy
    print("yes")

if "yt" in "python":   # Truthy
    print("yes")

if "quux" in ["foo", "bar", "baz"]:   # Falsy
    print("yes")
````

# <span style="font-family:Times New Roman">Grouping Statements: Indentation and Blocks</span>

<span style="font-family:Times New Roman">
So far, so good.  
But let’s say you want to evaluate a condition and then do more than one thing if it is true:

If the weather is nice, then I will:
- Mow the lawn  
- Weed the garden  
- Take the dog for a walk  

(If the weather isn’t nice, then I won’t do any of these things.)
</span>

---

<span style="font-family:Times New Roman">
In all the examples shown above, each <code>if &lt;expr&gt;:</code> has been followed by only a single statement.  
There needs to be some way to say **“If &lt;expr&gt; is true, do all of the following things.”**
</span>

<span style="font-family:Times New Roman">
The usual approach taken by most programming languages is to define a syntactic device that groups multiple statements into
one compound statement or block.  
A block is regarded syntactically as a single entity. When it is the target of an <code>if</code> statement,
and <code>&lt;expr&gt;</code> is true, then all the statements in the block are executed. If <code>&lt;expr&gt;</code> is false, then none of them are.
</span>

---

## <span style="font-family:Times New Roman">Python: It’s All About the Indentation</span>

<span style="font-family:Times New Roman">
Python follows a convention known as the <i>off-side rule</i>, a term coined by British computer scientist Peter J. Landin.  
(The term is taken from the offside law in association football.)  
Languages that adhere to the off-side rule define blocks by indentation.  
Python is one of a relatively small set of off-side rule languages.
</span>

<span style="font-family:Times New Roman">
Recall from the previous tutorial on Python program structure that indentation has special significance in a Python program.  
Now you know why: **indentation is used to define compound statements or blocks.**  
In a Python program, contiguous statements that are indented to the same level are considered to be part of the same block.
</span>

---

<span style="font-family:Times New Roman">
Thus, a compound <code>if</code> statement in Python looks like this:
</span>

```python
if <expr>:
    <statement>
    <statement>
    ...
    <statement>
<following_statement>



##### <span style="font-family:Times New Roman">
Consider the following Code:
</span>

````python
if 'foo' in ['bar', 'baz', 'qux']:
    print('Expression was true')
    print('Executing statement in suite')
    print('...')
    print('Done.')
print('After conditional')
````

<span style="font-family: 'Times New Roman', serif; font-size: 16px;">

### Understanding Blocks in Python

The four `print()` statements on lines **2 to 5** are indented to the same level as one another.  
They constitute the **block** that would be executed if the condition were true.  
But in this case, the condition is **false**, so all the statements in the block are skipped.  

After the end of the compound `if` statement has been reached (whether the statements in the block on lines 2 to 5 are executed or not), execution proceeds to the first statement having a **lesser indentation level**: the `print()` statement on **line 6**.

---

#### Important Notes:
- Blocks can be **nested** to arbitrary depth.  
- Each **indent** defines a new block.  
- Each **outdent** ends the preceding block.  

The resulting structure is **straightforward, consistent, and intuitive** in Python.  

</span>

<span style="font-family:Times New Roman">
Now look at this
</span>


````python
# Does line execute? Yes No
# --- --
if 'foo' in ['foo', 'bar', 'baz']: # x
    
    print('Outer condition is true') # x
    
    if 10 > 20: # x
    
        print('Inner condition 1') # x
    
    print('Between inner conditions') # x
    
    if 10 < 20: # x
    
        print('Inner condition 2') # x
    
    
    print('End of outer condition') # x
print('After outer condition') # x
````

# <span style="font-family: 'Times New Roman';">The `else` and `elif` Clauses</span>

<span style="font-family: 'Times New Roman';">
Now you know how to use an <code>if</code> statement to conditionally execute a single statement or a block of several statements.  
It’s time to find out what else you can do.
</span>  

<span style="font-family: 'Times New Roman';">
Sometimes, you want to evaluate a condition and take one path if it is true but specify an alternative path if it is not.  
This is accomplished with an <code>else</code> clause:
</span>  

```python
if <expr>:
    <statement(s)>
else:
    <statement(s)>
```
<span style="font-family: 'Times New Roman';">
If <code>&lt;expr&gt;</code> is true, the first suite is executed, and the second is skipped.  
If <code>&lt;expr&gt;</code> is false, the first suite is skipped and the second is executed.  

Either way, execution then resumes after the second suite.  
Both suites are defined by indentation, as described above.
</span>  

<span style="font-family: 'Times New Roman';">
In this example, <code>x</code> is less than 50, so the first suite (lines 4 to 5) is executed,  
and the second suite (lines 7 to 8) is skipped:
</span>  



<span style="font-family: 'Times New Roman';">
Here is an example demonstrating the use of the <code>else</code> clause:
</span>  


````python
x = 20

if x < 50:
    print("(first suite)")
    print("x is small")
else:
    print("(second suite)")
    print("x is large")
````

<span style="font-family: 'Times New Roman';">
Here, on the other hand, <code>x</code> is greater than 50, so the first suite is passed over, and the second suite executed:
</span>


````python
x = 100
if x < 50:
     print('(first suite)')
     print('x is small')
else:
     print('(second suite)')
     print('x is large')
````

<span style="font-family: 'Times New Roman';">
There is also syntax for branching execution based on several alternatives.  
For this, use one or more <code>elif</code> (short for <i>else if</i>) clauses.  

Python evaluates each <code>&lt;expr&gt;</code> in turn and executes the suite corresponding to the first that is true.  
If none of the expressions are true, and an <code>else</code> clause is specified, then its suite is executed:
</span>
``` python
if <expr>:
<statement(s)>
elif <expr>:
<statement(s)>
elif <expr>:
<statement(s)>
...
else:
<statement(s)

```
<span style="font-family: 'Times New Roman';">
An arbitrary number of <code>elif</code> clauses can be specified.  
The <code>else</code> clause is optional. If it is present, there can be only one, and it must be specified last:
</span>


````python
name = 'Joe'

if name == 'Fred':
    print('Hello Fred')
elif name == 'Xander':
    print('Hello Xander')
elif name == 'Joe':
    print('Hello Joe')
elif name == 'Arnold':
    print('Hello Arnold')
else:
    print("I don't know who you are!")
````

<span style="font-family: 'Times New Roman';">
At most, one of the code blocks specified will be executed.  
If an <code>else</code> clause isn’t included, and all the conditions are false, then none of the blocks will be executed.
</span>


````python
names = {
    'Fred': 'Hello Fred',
    'Xander': 'Hello Xander',
    'Joe': 'Hello Joe',
    'Arnold': 'Hello Arnold'
}

print(names.get('Joe', "I don't know who you are!"))
print(names.get('Rick', "I don't know who you are!"))
````

# One-Line if Statements

<span style="font-family: 'Times New Roman';">
It is customary to write <code>if &lt;expr&gt;</code> on one line and <code>&lt;statement&gt;</code> indented on the following line like this:
</span>

<pre>
if &lt;expr&gt;:
    &lt;statement&gt;
</pre>

<span style="font-family: 'Times New Roman';">
But it is permissible to write an entire <code>if</code> statement on one line. The following is functionally equivalent to the example above:
</span>

<pre>
if &lt;expr&gt;: &lt;statement_1&gt;; &lt;statement_2&gt;; ...; &lt;statement_n&gt;
</pre>

<span style="font-family: 'Times New Roman';">
But what does this mean? There are two possible interpretations:
</span>

<span style="font-family: 'Times New Roman';">
1. If <code>&lt;expr&gt;</code> is true, execute <code>&lt;statement_1&gt;</code>. Then, execute <code>&lt;statement_2&gt;</code> … <code>&lt;statement_n&gt;</code> unconditionally, irrespective of whether <code>&lt;expr&gt;</code> is true or not.  
2. If <code>&lt;expr&gt;</code> is true, execute all of <code>&lt;statement_1&gt;</code> … <code>&lt;statement_n&gt;</code>. Otherwise, don’t execute any of them.  
</span>

<span style="font-family: 'Times New Roman';">
Python takes the latter interpretation.  
The semicolon separating the <code>&lt;statements&gt;</code> has higher precedence than the colon following <code>&lt;expr&gt;</code>—in computer lingo, the semicolon is said to bind more tightly than the colon.  
</span>

<span style="font-family: 'Times New Roman';">
Thus, the <code>&lt;statements&gt;</code> are treated as a suite, and either all of them are executed, or none of them are.
</span>


````python
if 'f' in 'foo': print('1'); print('2'); print('3')

if 'z' in 'foo': print('1'); print('2'); print('3')
````

# Conditional Expressions (Python’s Ternary Operator)

Python supports one additional decision-making entity called a **conditional expression**.  
(It is also referred to as a **conditional operator** or **ternary operator** in various places in the Python documentation.)  

Conditional expressions were proposed for addition to the language in **PEP 308** and were **green-lighted by Guido in 2005**.  

In its simplest form, the syntax of the conditional expression is as follows:

```python
<expr1> if <condition> else <expr2>


````python
raining = False
print("Let's go to the", 'beach' if not raining else 'library')

raining = True
print("Let's go to the", 'beach' if not raining else 'library')

age = 12
s = 'minor' if age < 21 else 'adult'
print(s)

print('yes' if ('qux' in ['foo', 'bar', 'baz']) else 'no')
````

<span style="font-family: 'Times New Roman';">

### Conditional Expressions for Variable Assignment  

A common use of the conditional expression is to select variable assignment.  
For example, suppose you want to find the larger of two numbers.  

Of course, there is a built-in function, <code>max()</code>, that does just this (and more) that you could use.  
But suppose you want to write your own code from scratch.  

You could use a standard <code>if</code> statement with an <code>else</code> clause:
</span>


````python
a = 12
b = 6
if a > b:
    m = a
else:
    m = b
````

<span style="font-family: 'Times New Roman';">

### Conditional Expression (Short Form)

````python
 m = a if a > b else b
````

<span style="font-family: 'Times New Roman';">

### Conditional Expression in Larger Expressions

Remember that the conditional expression behaves like an **expression** syntactically.  
It can be used as part of a longer expression.  

The conditional expression has **lower precedence** than virtually all the other operators,  
so **parentheses are needed** to group it by itself.  

In the following example, the `+` operator binds more tightly than the conditional expression,  
so `1 + x` and `y + 2` are evaluated first, followed by the conditional expression.  
The parentheses in the second case are unnecessary and do not change the result:




````python
x = y = 40

z = 1 + x if x > y else y + 2
print(z)   

z = (1 + x) if x > y else (y + 2)
print(z)   
````

<span style="font-family: 'Times New Roman';">

### Grouping Conditional Expressions with Parentheses

If you want the conditional expression to be evaluated first, you need to **surround it with grouping parentheses**.  

In the next example, `(x if x > y else y)` is evaluated first.  
The result is `y`, which is **40**, so `z` is assigned `1 + 40 + 2 = 43`:
</span>

````python
x = y = 40
z = 1 + (x if x > y else y) + 2
z
````

<span style="font-family: 'Times New Roman';">

### Conditional Expressions and Short-Circuit Evaluation

If you are using a conditional expression as part of a larger expression,  
it is usually a good idea to use **grouping parentheses** for clarification,  
even if they are not strictly needed.  

Conditional expressions also use **short-circuit evaluation** like compound logical expressions.  
This means that portions of a conditional expression are **not evaluated** if they don’t need to be.  

In the expression:  
```python
<expr1> if <conditional_expr> else <expr2>


````python
print('foo' if True else 1/0)   
print(1/0 if False else 'bar')
````

# The Python `pass` Statement

Occasionally, you may find that you want to write what is called a **code stub**: a placeholder for where you will eventually put a block of code that you haven’t implemented yet.

In languages where **token delimiters** are used to define blocks, like the curly braces `{}` in **Perl** and **C**, empty delimiters can be used to define a code stub. For example, the following is legitimate Perl or C code:



````python
# This is not Python
if (x)
{  
}
````

<div style="font-family: 'Times New Roman', Times, serif;">

Here, the empty curly braces define an empty block. Perl or C will evaluate the expression `x`, and then even if it is true, quietly do nothing.

Because Python uses **indentation** instead of delimiters, it is not possible to specify an empty block. If you introduce an `if` statement with `if <expr>:`, something has to come after it, either on the same line or indented on the following line.

</div>


````python
if True:

print('foo')
````

<div style="font-family: 'Times New Roman', Times, serif;">

The Python `pass` statement solves this problem. It doesn’t change program behavior at all. It is used as a **placeholder** to keep the interpreter happy in any situation where a statement is syntactically required, but you don’t really want to do anything:





````python
if True:
    pass
print('foo')
````

### Voltage Level Classification 

    i.	A text file contains circuit voltage readings (one per line),
    ii.	Task: Read the file and classify each voltage as:
    a.	Low (< 110V)
    b.	Medium (110V–220V)
    c.	High (> 220V)

</span>

````python
with open("voltages.txt", "r") as file:
    readings = file.readlines()
    print(readings)
````

<span style="font-family: 'Times New Roman';">

    .readlines() → Reads all lines of the file into a list.

    Each line in the file becomes one string element in the list.

    Newline characters (\n) are kept at the end of each line.

</span>

````python
results = []  # store classified results

for line in readings:

    voltage = float(line.strip())  # convert string to float
    if voltage < 110:
        category = "Low"
    elif 110 <= voltage <= 220:
        category = "Medium"
    else:
        category = "High"
    results.append((voltage, category))
results
````
