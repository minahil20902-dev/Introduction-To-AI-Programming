1.	Scenario: Inventory system
A small shop keeps a list of product names: ["pencil", "notebook", "eraser", "marker"].
1.	Add two new products at the end.
2.	Insert a product at position 2.
3.	Explain why a list is appropriate here.


```python
# Task 1: Inventory System

print("=== Inventory System Task ===")

# Initial list of products
products = ["pencil", "notebook", "eraser", "marker"]
print(f"Initial inventory: {products}")

# 1.1 Add two new products at the end
products.append("sharpener")
products.append("ruler")
print(f"After adding two products: {products}")

# 1.2 Insert a product at position 2 (0-based index)
products.insert(2, "compass")
print(f"After inserting 'compass' at position 2: {products}")

# 1.3 Explain why a list is appropriate here
print("\n=== Explanation ===")
print("A list is appropriate here because:")
print("1. We need to maintain an ordered collection of products")
print("2. We need to frequently add new products (append/insert operations)")
print("3. We need to access products by their position (indexing)")
print("4. The inventory is dynamic and may change over time")
print("5. Lists allow duplicate items, which is realistic for inventory")
```

2.	Task: Data slicing & indexing
Given: temps = [23, 25, 19, 30, 21, 22]
1.	Access the first, last, and second-to-last items.
2.	Slice the list to get the middle four values.
3.	Using negative indexing, get the item “21”.



```python
# Task 2: Data Slicing & Indexing

print("=== Data Slicing & Indexing Task ===")

temps = [23, 25, 19, 30, 21, 22]
print(f"Original temperatures: {temps}")

# 2.1 Access the first, last, and second-to-last items
first_item = temps[0]
last_item = temps[-1]
second_last = temps[-2]
print(f"First item: {first_item}")
print(f"Last item: {last_item}")
print(f"Second to last: {second_last}")

# 2.2 Slice the list to get the middle four values
middle_four = temps[1:5]  # index 1, 2, 3, 4
print(f"Middle four values (index 1-4): {middle_four}")

# 2.3 Using negative indexing, get the item "21"
item_21 = temps[-2]  # 21 is second from the end
print(f"Item '21' using negative indexing: {item_21}")

print("\n=== Additional Slicing Examples ===")
# Different slicing techniques
print(f"First three: {temps[:3]}")
print(f"Last three: {temps[-3:]}")
print(f"Every other item: {temps[::2]}")
print(f"Reversed list: {temps[::-1]}")
```

3.	Task: Nested lists and heterogeneous data
Create a list representing three students: each student is itself a list containing [name, roll_number, grades_list]. Example: ["Ali", 101, [80, 90, 85]].
1.	Build for 3 students.
2.	Change one student’s second grade.
3.	Add a fourth grade to one student’s grades_list.



```python
# Task 3: Nested Lists and Heterogeneous Data

print("=== Nested Lists Task ===")

# 3.1 Build a list for 3 students
students = [
    ["Ali", 101, [80, 90, 85]],
    ["Sara", 102, [95, 88, 92]],
    ["Ahmed", 103, [75, 85, 78]]
]

print("Initial student records:")
for student in students:
    print(f"  {student}")

# 3.2 Change one student's second grade
# Change Sara's second grade (index 1 in grades list) from 88 to 91
students[1][2][1] = 91
print(f"\nAfter changing Sara's second grade:")
print(f"  Sara's new grades: {students[1][2]}")

# 3.3 Add a fourth grade to one student's grades_list
# Add a grade to Ahmed's grades
students[2][2].append(82)
print(f"\nAfter adding fourth grade to Ahmed:")
print(f"  Ahmed's new grades: {students[2][2]}")

print("\nFinal student records:")
for student in students:
    name, roll, grades = student
    avg_grade = sum(grades) / len(grades)
    print(f"  Name: {name}, Roll: {roll}, Grades: {grades}, Average: {avg_grade:.1f}")
```

6.	Task: List comprehensions & conversion
You have a tuple: values = (2, 4, 6, 8, 10).
1.	Convert the tuple to a list.
2.	Then use a list comprehension to produce a list where each value is squared.
3.	Filter this to include squared values greater than 30.
4.	Print the resulting list.



```python


print("=== List Comprehensions Task ===")


values = (2, 4, 6, 8, 10)
print(f"Original tuple: {values}")


values_list = list(values)
print(f"Converted to list: {values_list}")


squared_values = [x**2 for x in values_list]
print(f"Squared values: {squared_values}")


filtered_values = [x for x in squared_values if x > 30]
print(f"Values greater than 30: {filtered_values}")


print(f"\nFinal result: {filtered_values}")

print("\n=== More List Comprehension Examples ===")

evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"Even numbers 1-20: {evens}")


matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]
print(f"Flattened matrix: {flattened}")

# List comprehension with conditional expression
numbers = [1, 2, 3, 4, 5, 6]
parity = ["even" if x % 2 == 0 else "odd" for x in numbers]
print(f"Parity of numbers: {list(zip(numbers, parity))}")
```

1.	Scenario: Fixed configuration
You store configuration data for a device: device = ("device01", "192.168.1.10", 8080, True)
1.	Access each element by index.
2.	Attempt to change the port value (8080) to 9090; catch the error and explain why it happens.
3.	Convert the tuple into a list, modify the port to 9090, then convert back into a tuple.



```python
# Tuple Task 1: Fixed Configuration

print("=== Fixed Configuration Task ===")

# Store configuration data for a device
device = ("device01", "192.168.1.10", 8080, True)
print(f"Device configuration: {device}")

# 1.1 Access each element by index
print("\nAccessing elements by index:")
print(f"Device name: {device[0]}")
print(f"IP address: {device[1]}")
print(f"Port: {device[2]}")
print(f"Is active: {device[3]}")

# 1.2 Attempt to change the port value (8080) to 9090; catch the error
print("\nAttempting to modify tuple (will cause error):")
try:
    device[2] = 9090  # This will raise TypeError
    print("Modification successful")
except TypeError as e:
    print(f"Error occurred: {type(e).__name__}: {e}")
    print("Explanation: Tuples are immutable - their elements cannot be changed after creation")

# 1.3 Convert the tuple into a list, modify the port to 9090, then convert back
print("\nModifying through conversion:")
# Convert to list
device_list = list(device)
print(f"As list: {device_list}")

# Modify the port
device_list[2] = 9090
print(f"Modified list: {device_list}")

# Convert back to tuple
device_modified = tuple(device_list)
print(f"Modified tuple: {device_modified}")
```

4.	Task: Nested tuples & mixing
Create a tuple: data = ("UserA", (2023, 9, 15), ["task1", "task2"]).
1.	Access the date tuple and print year, month, day.
2.	Access the list of tasks and add a task to it. Explain how a mutable item in an immutable tuple can be modified.


```python

print("=== Nested Tuples Task ===")


data = ("UserA", (2023, 9, 15), ["task1", "task2"])
print(f"Original data structure: {data}")

date_tuple = data[1]
year, month, day = date_tuple
print(f"\nDate component: {date_tuple}")
print(f"Year: {year}, Month: {month}, Day: {day}")

tasks_list = data[2]
print(f"\nOriginal tasks: {tasks_list}")


tasks_list.append("task3")
print(f"After adding 'task3': {tasks_list}")


print(f"\nModified data structure: {data}")


print("\n=== Explanation ===")
print("How a mutable item in an immutable tuple can be modified:")
print("1. The tuple itself (data) is immutable - we cannot change which elements it contains")
print("2. However, the list inside the tuple (data[2]) is mutable")
print("3. We can modify the contents of that list without changing the tuple structure")
print("4. The tuple still references the same list object, but the list's contents have changed")
print("5. This is why tuples can contain mutable objects like lists, dictionaries, or sets")

print(f"\nList object ID before: {id(tasks_list)}")
tasks_list.append("task4")
print(f"List object ID after: {id(tasks_list)} (same object)")
print(f"Final tasks: {tasks_list}")
```

5.	Task: Packing & unpacking
Given: point = (10, 20, 30)
1.	Unpack into x, y, z.
2.	Swap x and z in one line using tuple unpacking.
3.	Pack x, y, z back into a tuple point2.
4.	Print both point and point2.


```python


print("=== Tuple Packing & Unpacking ===")


point = (10, 20, 30)
print(f"Original point: {point}")


x, y, z = point
print(f"\nUnpacked: x = {x}, y = {y}, z = {z}")


x, z = z, x
print(f"After swapping x and z: x = {x}, y = {y}, z = {z}")


point2 = (x, y, z)
print(f"New tuple point2: {point2}")


print(f"\nComparison:")
print(f"point: {point}")
print(f"point2: {point2}")

print("\n=== More Packing/Unpacking Examples ===")

values = (1, 2, 3, 4, 5)
first, *middle, last = values
print(f"Values: {values}")
print(f"First: {first}, Middle: {middle}, Last: {last}")

person = ("John", "Doe", 30, "Engineer")
first_name, last_name, age, _ = person
print(f"\nPerson: {person}")
print(f"First name: {first_name}, Last name: {last_name}, Age: {age}")

a, b = 5, 10
print(f"\nMultiple assignment: a = {a}, b = {b}")
```

6.	Scenario: Choosing tuple vs list
You need to store the months of the year as a fixed sequence.
1.	Define months as a tuple.
2.	Explain in comments why tuple is better than list here.
3.	Try to append a month and show the error.
4.	Convert to list if you really needed to change it.


```python

print("=== Choosing Tuple vs List ===")

months = ("January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December")

print("Months of the year:")
for i, month in enumerate(months, 1):
    print(f"  {i:2}. {month}")

print("\n=== Explanation in Code Comments ===")
print("""
# Why tuple is better than list for months:
# 1. The months of the year are fixed and never change
# 2. Using a tuple ensures the data cannot be accidentally modified
# 3. Tuples are more memory efficient than lists
# 4. Tuples can be used as dictionary keys (if needed)
# 5. It communicates to other developers that this is constant data
""")


print("Attempting to append a month to tuple (will cause error):")
try:
    months.append("Undecimber")  
except AttributeError as e:
    print(f"Error: {type(e).__name__}: {e}")
    print("Tuples don't have an append() method because they are immutable")


print("\nConverting to list for modification:")
months_list = list(months)
print(f"As list: {months_list}")

months_list.append("Undecimber")
print(f"After appending: {months_list}")


months_modified = tuple(months_list)
print(f"Back to tuple: {months_modified}")

print("\n=== Performance Consideration ===")
import sys
months_tuple = ("Jan", "Feb", "Mar")
months_list = ["Jan", "Feb", "Mar"]
print(f"Memory size of tuple: {sys.getsizeof(months_tuple)} bytes")
print(f"Memory size of list: {sys.getsizeof(months_list)} bytes")
print("Tuples generally use less memory than lists for the same data")
```

1.	Use in and not in to test membership.
2.	Concatenate nums + (15, 18).
3.	Multiply nums * 3.
4.	Use len(), min(), max(), sum() on the tuple


```python

print("=== Tuple Operators & Built-in Functions ===")

nums = (3, 6, 9, 12)
print(f"Original tuple: {nums}")

print("\nMembership tests:")
print(f"Is 9 in nums? {9 in nums}")
print(f"Is 5 in nums? {5 in nums}")
print(f"Is 10 not in nums? {10 not in nums}")

concatenated = nums + (15, 18)
print(f"\nConcatenated with (15, 18): {concatenated}")

multiplied = nums * 3
print(f"Multiplied by 3: {multiplied}")


print(f"\nBuilt-in functions:")
print(f"Length: {len(nums)}")
print(f"Minimum: {min(nums)}")
print(f"Maximum: {max(nums)}")
print(f"Sum: {sum(nums)}")

print("\n=== Tuple Comparison ===")

tuple1 = (1, 2, 3)
tuple2 = (1, 2, 4)
tuple3 = (1, 2, 3, 4)

print(f"tuple1 = {tuple1}")
print(f"tuple2 = {tuple2}")
print(f"tuple3 = {tuple3}")
print(f"\ntuple1 == tuple2: {tuple1 == tuple2}")
print(f"tuple1 < tuple2: {tuple1 < tuple2}") 
print(f"tuple1 < tuple3: {tuple1 < tuple3}")  
print(f"tuple2 > tuple3: {tuple2 > tuple3}") 

print("\n=== Immutability Demonstration ===")

print(f"Original nums id: {id(nums)}")
nums2 = nums + (15,)
print(f"After concatenation, new tuple id: {id(nums2)}")
print(f"Are they the same object? {nums is nums2}")
```
