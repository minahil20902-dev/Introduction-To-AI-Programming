3. map: map() applies a function to every item in a list.
Note: lambda is a short way to write a function in one line. Example: lambda x: x*2 means a function that doubles x.
i.	Convert a list of temperatures from Celsius to Fahrenheit using the formula F = (C × 9/5) + 32.
ii.	Convert a list of voltages in volts to kilovolts using V_kV = V / 1000.
Example:
nums = [1, 2, 3]
doubles = list(map(lambda x: x*2, nums))
print(doubles)
Output:
[2, 4, 6]


```python


print("=== Task 3.i: Celsius to Fahrenheit Conversion ===")
celsius_temps = [0, 20, 37, 100, -10]

# Using map with lambda
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

for c, f in zip(celsius_temps, fahrenheit_temps):
    print(f"{c}°C = {f:.1f}°F")

print("\n=== Task 3.ii: Volts to Kilovolts Conversion ===")
voltages = [1000, 2500, 11000, 22000, 33000]  # in Volts

# Using map with lambda
kilovolts = list(map(lambda v: v / 1000, voltages))

for v, kv in zip(voltages, kilovolts):
    print(f"{v} V = {kv} kV")
```

nums = [10, 20, 30]

17. round: round() rounds a number to the nearest integer or decimal places.
i.	Round a measured voltage to 2 decimal places.
ii.	Round frequency readings to the nearest integer.
Example:
x = 3.14159
print(round(x, 2))
Output:
3.14


```python
# Task 17: round() examples

print("=== Task 17.i: Round Voltage to 2 Decimal Places ===")
measured_voltage = 220.56789  # Volts

rounded_voltage = round(measured_voltage, 2)
print(f"Measured voltage: {measured_voltage} V")
print(f"Rounded to 2 decimal places: {rounded_voltage} V")

print("\n=== Task 17.ii: Round Frequency to Nearest Integer ===")
frequencies = [49.8, 50.2, 49.9, 50.1, 50.05]  # Hz

rounded_frequencies = [round(freq) for freq in frequencies]
print(f"Frequency readings: {frequencies} Hz")
print(f"Rounded to nearest integer: {rounded_frequencies} Hz")

print("\n=== Additional Examples ===")
# Different rounding scenarios
numbers_to_round = [
    (3.14159, 2),    # Pi to 2 decimal places
    (2.71828, 3),    # e to 3 decimal places
    (123.456, 1),    # To 1 decimal place
    (123.456, -1),   # To nearest 10
    (123.456, -2),   # To nearest 100
    (15.5, 0),       # Banker's rounding
    (14.5, 0),       # Banker's rounding
]

for num, ndigits in numbers_to_round:
    result = round(num, ndigits)
    print(f"round({num}, {ndigits}) = {result}")

# Special cases
print(f"\nround(2.675, 2) = {round(2.675, 2)}")  # Floating point precision issue
print(f"round(-3.14159, 2) = {round(-3.14159, 2)}")
```

18. divmod: divmod() returns the quotient and remainder of the division.
i.	87 optical fiber cables are drawn to provide high-speed internet for a university campus. The duct can contain 10 cables at most. Divide the total wires into bundles and find the number of completed filed ducts and how many more cables can be placed in the final duct.
ii.	Each group brings 42 LEDS for the DLD project. Divide the total number of LEDs into groups of 12, and find the groups and extra LEDs.
Example:
print(divmod(10, 3))
Output:
(3, 1)


```python
# Task 18: divmod() examples

print("=== Task 18.i: Optical Fiber Cable Bundling ===")
total_cables = 87
cables_per_duct = 10

complete_ducts, remaining_cables = divmod(total_cables, cables_per_duct)
print(f"Total cables: {total_cables}")
print(f"Cables per duct: {cables_per_duct}")
print(f"Complete ducts: {complete_ducts}")
print(f"Remaining cables: {remaining_cables}")
print(f"Cables that can be placed in final duct: {cables_per_duct - remaining_cables} more needed")

print("\n=== Task 18.ii: LED Distribution for DLD Project ===")
leds_per_group = 42
leds_per_subgroup = 12

complete_subgroups, extra_leds = divmod(leds_per_group, leds_per_subgroup)
print(f"LEDs per group: {leds_per_group}")
print(f"LEDs per subgroup: {leds_per_subgroup}")
print(f"Complete subgroups: {complete_subgroups}")
print(f"Extra LEDs: {extra_leds}")

print("\n=== Additional Examples ===")
# More divmod examples
examples = [
    (10, 3),    # From task description
    (17, 5),    # 17 divided by 5
    (100, 7),   # 100 divided by 7
    (256, 16),  # Exact division
    (5, 10),    # Dividend less than divisor
]

for dividend, divisor in examples:
    quotient, remainder = divmod(dividend, divisor)
    print(f"divmod({dividend}, {divisor}) = ({quotient}, {remainder})")
    print(f"  → {dividend} = {quotient} × {divisor} + {remainder}")
```

19. pow:pow() raises a number to the power of another.
i.	A list contains current values. Print the square of the current (I²) for power calculation.
Example:
print(pow(2, 3))
Output:
8


```python
# Task 19: pow() examples

print("=== Task 19.i: Square of Current Values ===")
current_values = [0.5, 1.0, 1.5, 2.0, 2.5]  # Amperes

# Calculate I² for each current value
current_squares = [pow(current, 2) for current in current_values]

print("Current (I) | I² (for power calculation)")
print("-" * 40)
for current, square in zip(current_values, current_squares):
    print(f"{current:^10} A | {square:^8.3f} A²")

print("\n=== Additional Examples ===")
# Basic pow examples
print(f"pow(2, 3) = {pow(2, 3)}")        # 2³ = 8
print(f"pow(5, 2) = {pow(5, 2)}")        # 5² = 25
print(f"pow(10, -1) = {pow(10, -1)}")    # 10⁻¹ = 0.1

# With three arguments (x**y % z)
print(f"\npow(4, 3, 5) = {pow(4, 3, 5)}")  # (4³) % 5 = 64 % 5 = 4

# Electrical calculations
print("\n=== Electrical Power Calculations ===")
resistances = [100, 220, 470, 1000]  # Ohms
current = 0.5  # Amperes

print(f"Power (P) = I² × R")
print(f"Current (I) = {current} A")
print("\nResistance (R) | Power (P)")
print("-" * 30)
for resistance in resistances:
    power = pow(current, 2) * resistance
    print(f"{resistance:^14} Ω | {power:^7.2f} W")
```
