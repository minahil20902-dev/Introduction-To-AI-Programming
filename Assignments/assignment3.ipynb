```python

import csv
print("\n\n=== ASSIGNMENT 3 ===")

loads = []
prices = []


with open('loads.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader) 
    
    for row in reader:
        load_value = float(row[0])
        loads.append(load_value)

print("Loads data loaded!")


with open('prices.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  
    
    for row in reader:
        price_value = float(row[0])
        prices.append(price_value)

print("Prices data loaded!")

print(f"\nLoads: {loads}")
print(f"Prices: {prices}")


def compute_bills(loads_list, prices_list):
    """
    Function to compute hourly bills and total bill
    """
    
    if len(loads_list) != len(prices_list):
        print("Error: Lists have different lengths!")
        return [], 0
    
    hourly_bills = []
    total_bill = 0
    
    
    for i in range(len(loads_list)):
        hourly = loads_list[i] * prices_list[i]
        hourly_bills.append(hourly)
        total_bill = total_bill + hourly  
    
    return hourly_bills, total_bill


hourly_bills, total_bill = compute_bills(loads, prices)

print("\n--- BILL CALCULATION ---")
print("Hour\tLoad\tPrice\tHourly Bill")
print("-" * 40)

for i in range(len(hourly_bills)):
    print(f"{i+1}\t{loads[i]}\t{prices[i]}\t{hourly_bills[i]:.2f}")

print("-" * 40)
print(f"TOTAL BILL: {total_bill:.2f}")
```
