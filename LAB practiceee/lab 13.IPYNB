Substation Fault Event Logging
A substation records fault events during operation.
Tasks:
	Create a fault log file
	Write initial fault event records
	Append future fault events as they occur


```python
def create_fault_log():
    header = "=== Substation Fault Event Log ==="
    
    initial_events = [
        "\nEvent 1: 2024-01-05 14:30",
        "Substation: SS-101",
        "Fault Type: Overcurrent",
        "Phase: Phase A",
        "Duration: 0.5 seconds",
        "Action: Auto-reclosed successfully",
        "Remarks: No equipment damage",
        "",
        "Event 2: 2024-01-07 09:15",
        "Substation: SS-102",
        "Fault Type: Earth Fault",
        "Phase: Phase B",
        "Duration: 2.0 seconds",
        "Action: Manual reset required",
        "Remarks: Insulator cleaning scheduled"
    ]
    
    with open('substation_faults.txt', 'w') as file:
        file.write(header + '\n')
        for event in initial_events:
            file.write(event + '\n')
    
    print("Initial fault log created")

def append_fault_event():
    new_event = [
        "\nEvent 3: 2024-01-10 16:45",
        "Substation: SS-103",
        "Fault Type: Voltage Dip",
        "Phase: All Phases",
        "Duration: 1.2 seconds",
        "Action: Voltage regulator activated",
        "Remarks: Caused by nearby construction"
    ]
    
    with open('substation_faults.txt', 'a') as file:
        for line in new_event:
            file.write(line + '\n')
    
    print("New fault event appended")

if __name__ == "__main__":
    create_fault_log()
    append_fault_event()
```

Task: Organize and Process CSV Data
Read the CSV file line by line and:
	Count the total number of recorded days
	Display each record with a serial number
The header must not be counted as a data record.


```python
import csv

def process_energy_data():
    try:
        with open('energy_data.csv', 'r') as file:
            reader = csv.reader(file)
            rows = list(reader)
        
        print("Original Energy Data:")
        print("-" * 50)
        
        data_rows = rows[1:]  # Skip header
        total_days = len(data_rows)
        
        print(f"Total number of recorded days: {total_days}\n")
        
        print("Data with serial numbers:")
        print("-" * 50)
        for i, row in enumerate(data_rows, 1):
            print(f"Record {i}: {row}")
    
    except FileNotFoundError:
        print("Creating sample energy_data.csv file...")
        
        sample_data = [
            ["Date", "Energy(kWh)", "Load(kW)", "Efficiency(%)"],
            ["2024-01-01", "1500", "125", "85"],
            ["2024-01-02", "1520", "128", "84"],
            ["2024-01-03", "1480", "123", "83"],
            ["2024-01-04", "1550", "130", "86"],
            ["2024-01-05", "1490", "124", "82"],
            ["2024-01-06", "1510", "126", "84"],
            ["2024-01-07", "1530", "129", "85"]
        ]
        
        with open('energy_data.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(sample_data)
        
        print("Sample file created. Please run again.")
        return

if __name__ == "__main__":
    process_energy_data()
```

Task: Write Organized Data to a New CSV File
Create a new file named organized_energy_data.csv.
Write the processed and formatted energy records into this file in CSV format.
The original file must remain unchanged.


```python
import csv

def organize_energy_data():
    try:
        with open('energy_data.csv', 'r') as file:
            reader = csv.reader(file)
            rows = list(reader)
        
        if len(rows) <= 1:
            print("No data to process")
            return
        
        header = rows[0]
        data_rows = rows[1:]
        
        organized_rows = [header]
        
        for i, row in enumerate(data_rows, 1):
            formatted_row = [f"Record_{i:03d}"] + row
            organized_rows.append(formatted_row)
        
        with open('organized_energy_data.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(organized_rows)
        
        print("Original file: energy_data.csv")
        print(f"Records processed: {len(data_rows)}")
        print("Organized data saved to: organized_energy_data.csv")
        
        print("\nSample of organized data:")
        print("-" * 50)
        for i in range(min(3, len(organized_rows))):
            print(organized_rows[i])
    
    except FileNotFoundError:
        print("Error: energy_data.csv not found")
        print("Please create the file or run process_csv.py first")

if __name__ == "__main__":
    organize_energy_data()
```
