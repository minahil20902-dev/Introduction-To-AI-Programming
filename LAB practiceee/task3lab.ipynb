1.	Task: Voltage Level Classification 
i.	A text file contains circuit voltage readings (one per line).
ii.	Task: Read the file and classify each voltage as:
a.	Low (< 110V)
b.	Medium (110V–220V)
c.	High (> 220V)
iii.	Save the classification results into a new file


```python

with open("voltage.txt", "r") as infile:
  
    voltages = [int(line.strip()) for line in infile]


results = []


for v in voltages:
    if v < 110:
        results.append(f"{v}V -> Low")
    elif 110 <= v <= 220:
        results.append(f"{v}V -> Medium")
    else:
        results.append(f"{v}V -> High")

with open("classified_voltages.txt", "w") as outfile:
    for result in results:
        outfile.write(result + "\n")

print("Voltage classification complete! Check 'classified_voltages.txt'")
```

3.	Task: Current Rating Check (Electrical Machines)
i.	A file has current values (in Amperes).
ii.	Task: Check for each line:
a.	If current < 5A → Safe
b.	If 5A ≤ current < 10A → Warning
c.	If ≥ 10A → Overload



```python
import os


directory = r"C:\Users\Dell\OneDrive\Desktop\ailab-VScode\lab3"

input_file = os.path.join(directory, "currents.txt")     
output_file = os.path.join(directory, "current_results.txt")  

with open(input_file, "r") as infile:
    currents = [line.strip() for line in infile if line.strip()]  

results = []
for c in currents:
    c = float(c) 
    if c < 5:
        results.append(f"{c}A -> Safe")
    elif c < 10:  
        results.append(f"{c}A -> Warning")
    else:         
        results.append(f"{c}A -> Overload")


with open(output_file, "w") as outfile:
    for result in results:
        outfile.write(result + "\n")


print("✅ Current Rating Check Results:")
for r in results:
    print(r)

print("\nReport complete! Check current_results.txt in your lab3 folder.")
```

4.	Task: File Extension Validator (General Engineering Reports)
i.	A directory has multiple engineering reports.
ii.	Task: Write a program to scan all files.
a.	If file ends with .pdf → Valid Report
b.	If file ends with .docx → Needs Conversion
c.	Otherwise → Invalid Format
iii.	Count and display the numbers of each type.


```python
import os

directory = r"C:\Users\Dell\OneDrive\Desktop\ailab-VScode\lab3"

valid_count = 0
conversion_count = 0
invalid_count = 0


for filename in os.listdir(directory):
    if filename.endswith(".pdf"):
        print(f"{filename} -> Valid Report")
        valid_count += 1
    elif filename.endswith(".docx"):
        print(f"{filename} -> Needs Conversion")
        conversion_count += 1
    else:
        print(f"{filename} -> Invalid Format")
        invalid_count += 1


print("\nSummary:")
print(f"Valid Reports (.pdf): {valid_count}")
print(f"Needs Conversion (.docx): {conversion_count}")
print(f"Invalid Format: {invalid_count}")
```

11.	Task:
Electrical systems run at a normal frequency of 50 or 60 Hz. Frequency outside this range may damage the equipment.
Check frequency value (Hz):
•	If < 40 → "Too Low"
•	Elif between 40–60 → "Normal Range"
•	Elif > 60 → "Too High"
•	Store frequency, check conditions step by step.


```python

try:
    freq = float(input("Enter frequency in Hz: "))
except ValueError:
    print("Please enter a numeric value (e.g. 50 or 60.0).")
else:
    if freq < 40:
        print(f"{freq} Hz -> Too Low")
    elif 40 <= freq <= 60:
        print(f"{freq} Hz -> Normal Range")
    else:
        print(f"{freq} Hz -> Too High")
```

Smart buildings adjust lighting automatically based on sensor lux values. We need to classify light levels.

•	Elif lux < 300 → "Dim Light – Lights ON low"
•	Elif lux < 600 → "Normal Brightness – Lights Balanced"
