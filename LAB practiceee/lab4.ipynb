2.	Ohm’s Law Calculator:
•   In your electrical lab, you need to calculate the voltage across a resistor.
•	Define a function that takes current (I) and resistance (R) and returns the voltage (V = I × R).
•	Call the function using positional, keyword, and default arguments.
•	Try mixing argument styles.
•	See what happens if you miss an argument


```python

def voltage(I: float = 1.0, R: float = 1.0) -> float:
    """
    Calculate voltage across a resistor using Ohm's Law.
    V = I * R

    Parameters
    ----------
    I : float, default = 1.0
        Current in amperes
    R : float, default = 1.0
        Resistance in ohms

    Returns
    -------
    float
        Voltage in volts
    """
    return I * R


print("1) Positional arguments:")
print("Voltage =", voltage(2, 5))   

print("\n2) Keyword arguments:")
print("Voltage =", voltage(R=10, I=0.5))  

print("\n3) Default arguments:")
print("No args (uses defaults I=1, R=1):", voltage())
print("Only R given (I=1 by default):", voltage(R=20))
print("Only I given (R=1 by default):", voltage(3))

print("\n4) Mixed styles:")
print("Voltage =", voltage(2, R=4))  
try:

    eval("voltage(I=2, 5)")
except SyntaxError:
    print("Error: SyntaxError → positional argument follows keyword argument")

print("\n5) Missing arguments without defaults:")
def voltage_strict(I, R):
    return I * R

try:
    print(voltage_strict(5))   
except TypeError as e:
    print("Error:", e)
```

A solar inverter monitors the battery’s state of charge (SOC).
•	Write a function charge_from_solar(SOC) that:

o	Prints "Battery full, no action" if SOC > 90%.
o	Prints "Battery charging, SOC = ---%" otherwise.
•	Add docstrings and function annotations.

```python
def charge_from_solar(SOC: float) -> None:
    """
    Decide solar battery charging action based on State of Charge (SOC).

    Parameters
    ----------
    SOC : float
        State of Charge of the battery in percentage (0–100).

    Behavior
    --------
    - If SOC < 20 → prints "Charging".
    - If SOC > 90 → prints "Battery full, no action".
    - Otherwise   → prints "Battery charging, SOC = ---%".
    """
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print(f"Battery charging, SOC = {SOC:.1f}%")

print("Test 1: SOC = 10")
charge_from_solar(10)

print("\nTest 2: SOC = 95")
charge_from_solar(95)

print("\nTest 3: SOC = 55.5")
charge_from_solar(55.5)


print("\nDocstring:")
print(charge_from_solar.__doc__)
```

4.	Resistance, Voltage, and Current Lists:
You are building a program to analyze five resistor values in a circuit.
•	Function 1: resistance():  return a list of 5 resistances.
•	Function 2: voltage():  return voltages across each resistor.
•	Function 3: current():  take both lists (Resistance & Voltage) and calculate currents.
•	Display all lists nicely in the main program.
•	Try two versions:
o	Return as a tuple.
o	Return as a dictionary with keys like "current_1".


```python

def resistance() -> list[float]:
    """
    Returns a list of 5 resistor values (ohms).
    """
    return [10, 20, 30, 40, 50]



def voltage() -> list[float]:
    """
    Returns a list of 5 voltages (volts) across the resistors.
    """
    return [5, 10, 15, 20, 25]


def current_tuple(R: list[float], V: list[float]) -> tuple:
    """
    Calculate currents (I = V / R) and return as a tuple.
    """
    return tuple(v / r for v, r in zip(V, R))



def current_dict(R: list[float], V: list[float]) -> dict:
    """
    Calculate currents (I = V / R) and return as a dictionary.
    Keys: "current_1", "current_2", etc.
    """
    return {f"current_{i+1}": v / r for i, (v, r) in enumerate(zip(V, R))}



if __name__ == "__main__":
    R = resistance()
    V = voltage()

    print("Resistances (ohms):", R)
    print("Voltages (volts):  ", V)


    I_tuple = current_tuple(R, V)
    print("\nCurrents as tuple (A):", I_tuple)

    I_dict = current_dict(R, V)
    print("\nCurrents as dictionary (A):")
    for key, value in I_dict.items():
        print(f"  {key}: {value:.2f}")
```

5.	Series Resistance:
In circuit design, you often calculate series resistance.
•	Write a function series() that takes a list or tuple of resistances.
•	Return the total resistance.
•	Add docstrings and annotations.
•	Print the function’s __annotations__.


```python
def series(resistors: list[float] | tuple[float, ...]) -> float:
    """
    Calculate the total resistance for resistors connected in series.

    Parameters:
        resistors (list[float] | tuple[float, ...]): 
            A list or tuple containing resistance values in ohms.

    Returns:
        float: The total resistance (sum of all resistors) in ohms.
    """
    return sum(resistors)


resistor_values = [10, 20, 30, 40, 50]
total = series(resistor_values)

print("Resistors (Ω):", resistor_values)
print("Total Series Resistance (Ω):", total)

print("\nFunction Annotations:")
print(series.__annotations__)
```

6.	Parallel Resistance:
Now, calculate parallel resistance.
•	Write a function parallel() that takes a list or tuple of resistances.
•	Return the equivalent resistance.
•	Add docstrings and annotations.
•	Print the function’s __annotations__.


```python
def parallel(resistors: list[float] | tuple[float, ...]) -> float:
    """
    Calculate the equivalent resistance for resistors connected in parallel.

    Parameters:
        resistors (list[float] | tuple[float, ...]):
            A list or tuple containing resistance values in ohms.

    Returns:
        float: The total equivalent resistance (in ohms).
    """
 
    if 0 in resistors:
        raise ValueError("Resistance cannot be zero in a parallel circuit.")
    
    reciprocal_sum = sum(1 / r for r in resistors)
    return 1 / reciprocal_sum



resistor_values = [10, 20, 30, 40, 50]
equiv_res = parallel(resistor_values)

print("Resistors (Ω):", resistor_values)
print("Equivalent Parallel Resistance (Ω):", round(equiv_res, 2))


print("\nFunction Annotations:")
print(parallel.__annotations__)
```

7.	Series Resistance with *args:

•	Define series(*args) using tuple packing.
•	Pass any number of resistances.

•	Add docstrings and annotations.
•	Print __annotations__.



```python
def series(*args: float) -> float:
    """
    Calculate the total resistance for resistors connected in series.

    Parameters:
        *args (float):
            Any number of resistance values in ohms.

    Returns:
        float: The total resistance (sum of all resistors) in ohms.
    """
    return sum(args)



total_resistance = series(10, 20, 30, 40, 50)

print("Resistances (Ω):", (10, 20, 30, 40, 50))
print("Total Series Resistance (Ω):", total_resistance)


print("\nFunction Annotations:")
print(series.__annotations__)
```

8.	Parallel Resistance with *args:
Do the same for parallel resistance.
•	Define parallel(*args) using tuple packing.
•	Pass any number of resistances.
•	Return the equivalent resistance.
•	Add docstrings and annotations.
•	Print __annotations__.


```python
def parallel(*args: float) -> float:
    """
    Calculate the equivalent resistance for resistors connected in parallel.

    Parameters:
        *args (float):
            Any number of resistance values in ohms.

    Returns:
        float: The total equivalent resistance in ohms.
    """
    
    if not args:
        raise ValueError("At least one resistance value must be provided.")
    

    if any(r == 0 for r in args):
        raise ValueError("Resistance value cannot be zero in a parallel circuit.")

    
    reciprocal_sum = sum(1 / r for r in args)
    return 1 / reciprocal_sum



equiv_res = parallel(10, 20, 30, 40, 50)

print("Resistances (Ω):", (10, 20, 30, 40, 50))
print("Equivalent Parallel Resistance (Ω):", round(equiv_res, 2))

print("\nFunction Annotations:")
print(parallel.__annotations__)
```

2. Using the chain rule,compute the derivate of z with respect to x:
  dz/dx=dz/dy.dy/dx

```python
import math



def dz_dx(x: float) -> float:
    """Compute the derivative of z with respect to x using the chain rule."""
    y = x ** 2
    dzdx = math.cos(y) * (2 * x)
    return dzdx


for val in [0.5, 1.0, 2.0]:
    print(f"x = {val}, dz/dx = {dz_dx(val):.5f}")
```
