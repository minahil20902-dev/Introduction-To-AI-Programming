1)  Smart Home Energy Meter:-
A smart home meter logs real-time energy usage
{    "time": "18:32",
    "grid_power": 720,
    "solar_power": 340,
    "battery_soc": 67,
    "appliance_status": {"AC": "OFF", "Fridge": "ON", "Motor": "OFF"}}
1.	Print the "grid_power" value.
2.	Add a new appliance "TV": "ON" inside "appliance_status".
3.	Change "battery_soc" to 72.
4.	Check if "solar_power" exists in the dictionary.
5.	Delete the "Motor" key from "appliance_status".


```python
energy_meter = {
    "time": "18:32",
    "grid_power": 720,
    "solar_power": 340,
    "battery_soc": 67,
    "appliance_status": {
        "AC": "OFF",
        "Fridge": "ON",
        "Motor": "OFF"
    }
}

print(energy_meter["grid_power"])

energy_meter["appliance_status"]["TV"] = "ON"

energy_meter["battery_soc"] = 72

print("solar_power" in energy_meter)

del energy_meter["appliance_status"]["Motor"]
```

2)  University Student Record
A system stores basic info about a student:
{  "name": "Ali Khan",
    "roll_no": "EE-22-045",
    "gpa": 3.41,
    "courses": ["Circuit Analysis", "Linear Algebra", "Python"],
    "attendance": {"Circuit Analysis": 92, "Linear Algebra": 88, "Python": 96}}
1.	Print the studentâ€™s GPA.
2.	Add a new course "Digital Logic" with attendance 90.
3.	Update "Linear Algebra" attendance to 91.
4.	Check if the key "gpa" exists


```python
student = {
    "name": "Ali Khan",
    "roll_no": "EE-22-045",
    "gpa": 3.41,
    "courses": ["Circuit Analysis", "Linear Algebra", "Python"],
    "attendance": {
        "Circuit Analysis": 92,
        "Linear Algebra": 88,
        "Python": 96
    }
}

print(student["gpa"])

student["courses"].append("Digital Logic")
student["attendance"]["Digital Logic"] = 90

student["attendance"]["Linear Algebra"] = 91

print("gpa" in student)
```

3)  Weather Monitoring Station
Hourly weather data is stored as a dictionary:
{   "hour": 14,
    "temperature": 33.8,
    "humidity": 52,
    "wind": {"speed": 12, "direction": "NE"},
    "alerts": []}

1.	Print the wind direction.
2.	Add a new alert "High Temperature".
3.	Change the wind speed to 15.
4.	Use .keys() to list all top-level keys.
5.	Remove "humidity" using .pop().


```python
weather_data = {
    "hour": 14,
    "temperature": 33.8,
    "humidity": 52,
    "wind": {
        "speed": 12,
        "direction": "NE"
    },
    "alerts": []
}

print(weather_data["wind"]["direction"])

weather_data["alerts"].append("High Temperature")

weather_data["wind"]["speed"] = 15

print(weather_data.keys())

weather_data.pop("humidity")
```

4)  Inventory System for Electronics Shop
A shop tracks an item in its inventory:
{
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {"voltage": "12/24V", "current": "20A", "brand": "SolarTech"}
}
1.	Print the brand of the controller.
2.	Reduce the stock by 2 and update the dictionary.
3.	Add a new feature "warranty": "1 year".
4.	Check if "price" exists in the dictionary.
5.	Merge another dictionary {"discount": 10} into the inventory item.


```python
inventory = {
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {
        "voltage": "12/24V",
        "current": "20A",
        "brand": "SolarTech"
    }
}

print(inventory["features"]["brand"])

inventory["stock"] -= 2

inventory["features"]["warranty"] = "1 year"

print("price" in inventory)

inventory.update({"discount": 10})
```

5)  Hospital Patient Monitoring System
Patient vitals are recorded every 10 minutes:
{
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}
}
1)	Print the oxygen level.
2)	Add a new medication "antibiotic": "6PM".
3)	Update heart rate to 78.
4)	Get the medication schedule for "insulin" using .get() with default "Not Prescribed".
5)	Use .items() to print all key-value pairs.


```python
patient = {
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {
        "aspirin": "8AM",
        "vitamin_D": "12PM"
    }
}

print(patient["oxygen_level"])

patient["medications"]["antibiotic"] = "6PM"

patient["heart_rate"] = 78

print(patient["medications"].get("insulin", "Not Prescribed"))

print(patient.items())
```
